```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('/content/AirQualityUCI.xlsx')
```

```python
print(data.isnull().sum())
```

```python
print("First few rows of the dataset:")
print(data.head())
```

```python
print(data['Date'].describe())
```

```python
print("\nBasic statistics of the dataset:")
print(data.describe())
```

```python
# Plot histograms for each column
print("\nPlotting histograms for each column...")
data.hist(bins=30, figsize=(18, 14))
plt.tight_layout()
plt.show()
```

```python
print("\nPlotting pairplot for key variables...")
sns.pairplot(data[['CO(GT)', 'NMHC(GT)', 'C6H6(GT)', 'NOx(GT)', 'NO2(GT)', 'T', 'RH', 'AH']])
plt.show()
```

```python
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = data[numerical_columns].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Columns')
plt.show()
```
